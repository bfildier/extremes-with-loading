# Extreme updrafts at each level

```python
wshape = data3D.wa.shape
i_z = 10
wa_i = data3D.wa.isel(z=i_z)

dist_wa_i = Distribution('wa_%d'%i_z,nbpd=10,nd=6)
dist_wa_i.computeDistribution(wa_i.data.ravel())
```

```python
fig,ax = plt.subplots(figsize=(5,5))
setFrame(ax,rankmax=99.9999)
showData(ax,dist_wa_i.ranks,dist_wa_i.percentiles,rankmax=99.9999)

ax.set_ylim((-1,10))
ax.set_ylabel(r'$w_i^e$ (m/s)')
ax.set_xlabel(r'rank $e$ (%)',labelpad=25)
```

```python
# Compute distribution at all levels

Nz = wshape[1]

dist_all = []
for i_z in range(Nz):
    
    wa_i = data3D.wa.isel(z=i_z)
    dist_wa_i = Distribution('wa_%d'%i_z,nbpd=10,nppb=4,nd=6)
    dist_wa_i.computeDistribution(wa_i.data.ravel())
    dist_all.append(dist_wa_i)
```

```python
norm = matplotlib.colors.Normalize(vmin=0, vmax=Nz-1)
cmap = matplotlib.cm.get_cmap('Spectral')

fig,ax = plt.subplots(figsize=(5,5))
setFrame(ax,rankmax=99.9999)

for i_z in range(Nz):
    
    dist = dist_all[i_z]
    showData(ax,dist.ranks,dist.percentiles,rankmax=99.9999,c=cmap(norm(i_z)))

ax.set_ylim((-1,30))
ax.set_ylabel(r'$w_i^e$ (m/s)')
ax.set_xlabel(r'rank $e$ (%)',labelpad=25)
```

```python
# Extract profile of w^e for specific percentiles

ranks_all = [99,99.5,99.9,99.95,99.99,99.995,99.999]
Nc = len(ranks_all)
i_ranks_all = [dist_all[10].indexOfRank(ranks_all[j]) for j in range(Nc)]
ranks_i_all = [dist_all[10].ranks[i_e] for i_e in i_ranks_all] # check we are close to target ranks

wa_ranks_all = []

for i_z in range(Nz):
    
    wa_i_all = [dist_all[i_z].percentiles[i_e] for i_e in i_ranks_all] # get updraft speed
    wa_ranks_all.append(wa_i_all)
    
wa_ranks = np.vstack(wa_ranks_all)
    
```

```python
fig,ax = plt.subplots(figsize=(5,5))

norm = matplotlib.colors.Normalize(vmin=0, vmax=Nc-1)
cmap = matplotlib.cm.get_cmap('Spectral_r')

z = data3D.z.data/1e3

for i_c in range(Nc):
    
    ax.plot(wa_ranks[:,i_c],z,c=cmap(norm(i_c)))
    
ax.set_xlabel('updraft speed (m/s)')
ax.set_ylabel('z (km)')
```

```python
fig,axs = plt.subplots(ncols=2,figsize=(10,5))

norm = matplotlib.colors.Normalize(vmin=0, vmax=Nc-1)
cmap = matplotlib.cm.get_cmap('Spectral_r')

z = data3D.z.data/1e3
pa = data3D.pa[0,:,0,0]

# Temperature profile
ax = axs[0]

ax.plot(ta_mean,ta.z/1e3,c='k')
ax.set_xlabel('mean T (K)')
ax.set_ylabel('z (km)')

# extreme updrafts
ax = axs[1]

for i_c in range(Nc):
    
    ax.plot(wa_ranks[:,i_c],z,c=cmap(norm(i_c)),label=r'%2.3fth'%ranks_all[i_c])
    
ax.set_xlabel('updraft speed (m/s)')
ax.set_ylabel('z (km)')
ax.legend()

plt.savefig(os.path.join('../figures','meanT_w_ext_SAM_small300.pdf'),bbox_inches='tight')
```

# Export dataset

```python
# Initialize dataset with pre-computed MCS characteristics

ds = xr.Dataset(
    data_vars=dict(
        ta_mean=(['z'],ta_mean),
        wa_ranks=(['z','e'],wa_ranks),
    ),
    coords=dict(
        z=z, # height, m
        e=ranks_all, # extreme percentile rank, %
    ),
    attrs=dict(description="Some profiles from the RCEMIP SAM_small300 simulation"),
)

# attributes
ds['ta_mean'].attrs['description'] = 'Mean temperature profile'
ds['ta_mean'].attrs['units'] = 'K'
ds['wa_ranks'].attrs['description'] = 'Extreme updraft speeds at specific percentile ranks'
ds['wa_ranks'].attrs['units'] = 'm/s'
ds['z'].attrs['description'] = 'vertical coordinate'
ds['z'].attrs['units'] = 'km'
ds['e'].attrs['description'] = 'extreme percentile ranks: 99th, 99.5th, 99.9th, 99.95th, 99.99th, 99.995th, 99.999th'
ds['e'].attrs['units'] = '%'

ds.to_netcdf(os.path.join('../output','ta_wa_ext_SAM_small300.nc'))
```
